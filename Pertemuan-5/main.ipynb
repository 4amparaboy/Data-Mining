```python
# Mengimport library yang dibutuhkan
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Membaca data dari CSV
data = pd.read_csv('data.csv')

# Menampilkan dataFrame
print(data)
```

```python
# Memisahkan variabel independen (X) / "Anggaran Iklan"
# dan variabel dependen (y) / "Penjualan Unit"
X = data['Anggaran Iklan (Rp)'].values.reshape(-1, 1)
y = data['Penjualan (Unit)'].values.reshape(-1, 1)
```

```python

# Membuat model menggunakan LinearRegression
model = LinearRegression()
model.fit(X, y)

# Koefisien Regresi (m)
m = model.coef_[0][0]

# Intercept c
c = model.intercept_[0]
```

```python

# Membuat plot
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title('Regresi Linier: Penjualan vs Anggaran Iklan')
plt.xlabel('Anggaran Iklan (Rp)')
plt.ylabel('Penjualan (Unit)')
plt.show()

# Menampilkan slope (m) dan intercept (c)
print("Koefisien Regresi (m):", m)
print("Intercept (c):", c)

# Prediksi penjualan jika anggaran 300000
anggaran_prediksi = 300000
# penjualan_prediksi = m * anggaran_prediksi + c
penjualan_prediksi = model.predict([[anggaran_prediksi]])
print(f"Prediksi penjualan jika anggaran Rp 300.000: {penjualan_prediksi[0][0]}")
```
